```python
from pathlib import Path

def clean_line(line: str):
    return " ".join([word.strip() for word in line.strip().split(" ")])

def clean_file(source: Path, dest: Path):
    if not source.exists():
        raise FileNotFoundError(f"{source} does not exist")

    with (
        source.open("r", encoding="utf-8", errors="ignore") as fin,
        dest.open("w", encoding="utf-8") as fout
    ):
        for line in fin:
            fout.write(clean_line(line) + "\n")
```
